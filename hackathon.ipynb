```python
import re    # regular expression
from nltk.corpus import stopwords

# set of stop words
stop = set(stopwords.words('english')) 

def remove_stop(word):
    if word not in stop:
        return word
    return ''
        
def lower_text(sentence):
    sentence=sentence.lower()
    return sentence

def cleanhtml(sentence):
    cleanr = re.compile('<.*?>')  # any thing in angular brackets
    cleantext = re.sub(cleanr, 'link', sentence) # replace with
    return cleantext

def cleanpunc(sentence):
    cleaned = re.sub(r'[?|!|\'|"|#]',r'', sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return cleaned
```

```python
for i in range(len(df['lst2Tite'])):
#     print(df['lst2Tite'].iloc[i])
    df['lst2Tite'].iloc[i] = lower_text(df['lst2Tite'].iloc[i])
    df['lst2Tite'].iloc[i] = cleanpunc(df['lst2Tite'].iloc[i])
    df['lst2Tite'].iloc[i] = cleanhtml(df['lst2Tite'].iloc[i])
```
